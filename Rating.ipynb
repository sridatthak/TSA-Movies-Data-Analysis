```python
#Rating Analysis
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x="rating", y="score", data=df)
```

```python
#All the different ratings didn't have much deviation between them. However, the "Approved" rating stood out as the rating with the least average score. Rating X was the most successful in terms of average score just by a slight amount
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x="rating", y="votes", data=df)
```

```python
#This graph was very different from the last one. This is because the success of the movie is now being measured based on votes. PG-13 had the maximum amount of average votes and G was right behind it. The least number of votes by a huge amount was in the Approved column. 
```

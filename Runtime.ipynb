```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
df = pd.read_csv("movies_cleaned.csv")
```

```python

# Set plot style
sns.set_style("whitegrid")
```

```python
# 1. Distribution of Movie Runtime
plt.figure(figsize=(12, 6))
sns.histplot(df["runtime"], bins=30, kde=True, color="blue")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Count of Movies")
plt.title("Distribution of Movie Runtime")
plt.show()
```

```python
#Most movies have a runtime between 80 to 120 minutes, with a peak around 100 minutes.
#Very few movies exceed 150 minutes, indicating that extremely long films are uncommon.
```

```python
# 2. Runtime vs. IMDB Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["runtime"], y=df["score"], alpha=0.6, color="red")
plt.xlabel("Runtime (minutes)")
plt.ylabel("IMDB Score")
plt.title("Runtime vs. IMDB Score")
plt.show()
```

```python
# There is no strong correlation between runtime and IMDB score.
# Some longer movies (above 120 minutes) have high ratings, but highly rated films exist across all runtime ranges.
```

```python

# 4. Runtime vs. Gross Revenue
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["runtime"], y=df["gross"], alpha=0.6, color="purple")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Gross Revenue (in dollars)")
plt.title("Runtime vs. Gross Revenue")
plt.yscale("log")  # Log scale for better visualization
plt.show()
```

```python
# Movies with longer runtimes tend to earn more, but the relationship is not absolute.
# There are short movies with high earnings, suggesting that factors like genre, cast, and marketing play crucial roles.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "movies_cleaned.csv"
df = pd.read_csv(file_path)
```

```python
# Set plot style
sns.set_style("whitegrid")
```

```python
# 9. Distribution of Budget
plt.figure(figsize=(12, 6))
sns.histplot(df["budget"], bins=30, kde=True, color="brown")
plt.xlabel("Budget (in dollars)")
plt.ylabel("Count of Movies")
plt.title("Distribution of Movie Budgets")
plt.yscale("log")  # Log scale for better visualization
plt.show()
# Conclusion: Most movies have relatively low budgets, but some big productions have extremely high budgets.
```

```python
# 10. Budget vs. IMDB Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["score"], y=df["budget"], alpha=0.6, color="purple")
plt.xlabel("IMDB Score")
plt.ylabel("Budget (in dollars)")
plt.title("Budget vs. IMDB Score")
plt.yscale("log")  # Log scale for better visualization
plt.show()
# Conclusion: Highly rated movies exist across different budget ranges, but some of the highest-rated films have moderate budgets.
```

```python
#Budget vs. Gross Revenue
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["budget"], y=df["gross"], alpha=0.6, color="red")
plt.xlabel("Budget (in dollars)")
plt.ylabel("Gross Revenue (in dollars)")
plt.title("Budget vs. Gross Revenue")
plt.yscale("log") # Log scale for better visualization
plt.xscale("log") # Log scale for better visualization
plt.show()
# Conclusion: Higher budget movies tend to earn more, but there are outliers where low-budget movies achieved high revenue.
```
